importing the required tensorflow modules

```python
import keras
import numpy as np
```

```python
base_model = keras.applications.ResNet152(weights='imagenet',
include_top=False,
input_shape=(224, 224, 3))
```

```python
base_model.summary()
```

```python
from keras import models, regularizers, layers, optimizers, losses, metrics
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = models.Sequential()
model.add(base_model)
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dropout(0.5))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(2, activation='sigmoid', kernel_regularizer=regularizers.l2(0.1)))
```

```python
model.summary()
```

```python
# freezing the modle for fine-tuning the last layer
for layer in base_model.layers[:]:
   layer.trainable = False

print('base_model is now NOT trainable')
```

For fun purposes, let's explore the parts of the resnet152

```python
for i, layer in enumerate(base_model.layers):
   print(i, layer.name, layer.trainable)
```

```python
# Compiling the model

model.compile(optimizer=optimizers.Adam(learning_rate=0.00001, beta_1=0.9, beta_2=0.999),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```
